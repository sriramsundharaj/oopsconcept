```python
class MenuItems:

    def __init__(self,name,price,category):
        self.name = name
        self.price = price
        self.category = category


    def __str__(self):
        return f"{self.name} ${self.price} {self.category}"


class Order:

    def __init__(self):
        self.items = []

    def add_items(self,menu_item,quantity):
        self.items.append((menu_item,quantity))
        print(f"Added {quantity} x {menu_item.name} to the order.")

    def display_order(self):
        if not self.items:
            print('Your order is empty!')
        print("\nCurrent order :")
        total = 0
        for item,quantity in self.items:
            print(f"{quantity} X {item.name} = {quantity * item.price}")
            total += item.price * quantity
        print(f"Total :${total}")

    def calculate_total(self):
        return sum(item.price * quantity for item,quantity in self.items)


class Restuarant:

    def __init__(self):
        self.menu = []


    def added_menu_items(self,name,price,category):
        self.menu.append(MenuItems(name,price,category))

    def display_menu(self):
        print("\nMenu :")
        for index,menu in enumerate(self.menu,start=1):
            print(f"{index}. {menu}")

    def take_order(self):
        order = Order()
        while True:
            self.display_menu()
            choice = input("Enter the menu item number to add to your order (or 'done' to finish): ")
            if choice.lower() == 'done':
                return order
            try:
                index = int(choice) - 1
                if 0 <= index < len(self.menu):
                    quantity = int(input(f"Enter quantity for {self.menu[index].name}: "))
                    order.add_items(self.menu[index],quantity)
                else:
                    print('Invalid number!!')
            except ValueError:
                print("Invalid input. Please enter a number.")




my_restuarant = Restuarant()
my_restuarant.added_menu_items('pasta',10,'breakfast')
my_restuarant.added_menu_items('pizza',10,'launch')
my_restuarant.added_menu_items('noodles',10,'dinner')
my_order = my_restuarant.take_order()
print(f"Total : {my_order.calculate_total()}")
```
